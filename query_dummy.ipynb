```python
import ipyvuetify as v
from IPython.display import display


# Create banner
banner = v.Card(
    class_='mb-4',
    elevation=3,
    children=[
        v.CardTitle(
            class_='white--text text-h4 justify-center',
            style_='background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); padding: 30px;',
            children=['Example Notebook']
        )
    ]
)

# Create message card
message = v.Card(
    class_='pa-6',
    elevation=2,
    children=[
        v.CardText(
            class_='text-h5 text-center',
            children=['Example notebook for the git presentation']
        )
    ]
)

# Create layout
layout = v.Container(
    fluid=True,
    class_='pa-4',
    style_='max-width: 800px;',
    children=[
        banner,
        message
    ]
)

# Display
display(layout)
```
